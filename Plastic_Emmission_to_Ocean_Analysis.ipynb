```python
rivers = df['Country'].value_counts()
River_Total = pd.DataFrame(rivers)
River_Total.head(2)
```

```python
River_Total.rename(columns={'Country':'river_total'},inplace=True)
River_Total
```

```python
River_Total['count'].sum()
```

```python
fig = px.bar(River_Total, x=River_Total.index, y="count")
fig.show()
```
